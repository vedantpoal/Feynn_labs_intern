Finding the outliers in the dataset

```python
def remove_outlier(col):
  sorted(col)
  Q1,Q3=col.quantile([0.25,0.75])
  IQR=Q3-Q1
  lower_range=Q1-(1.5*IQR)
  upper_range=Q3+(1.5*IQR)
  return lower_range,upper_range
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.boxplot(column=['AccelSec'])
plt.show()
```

```python
low,up=remove_outlier(df['AccelSec'])
df['AccelSec']=np.where(df['AccelSec']>up,up,df['AccelSec'])
df['AccelSec']=np.where(df['AccelSec']<low,low,df['AccelSec'])
```

```python
df['PowerTrain']=my_label.fit_transform(df['PowerTrain'])
```

```python
# Filteration of models related to Powertrain
df2 = df[df['PowerTrain']>=1] # Considering only Forward wheel drive and rear wheel drive

plt.figure(figsize=(15,5))
plt.title("Powertrain",fontsize=25)
plt.grid(axis='y')
plt.ylim(bottom=0,top=3)
plt.xticks(rotation = 90)
plt.xlabel("Model Name",fontsize=20)
plt.ylabel("Powertrain",fontsize=20)
plt.plot(df2['Model'],df2['PowerTrain'],label='Range',c='r')
```

```python
grouped = df2.groupby('BodyStyle')['Price'].mean()

# Plotting the bar graph
plt.bar(grouped.index, grouped.values, color='skyblue')
plt.xlabel('Body Style')
plt.ylabel('Average Price')
plt.title('Average Price by Car Body Style')
plt.xticks(rotation=45)  # Rotating x-axis labels if needed
plt.show()
```

```python
# Filteration of models related to Age
df3 = df[df['Age']>=30] # Age above 30

plt.figure(figsize=(15,5))
plt.title("Age",fontsize=25)
plt.grid(axis='y')
plt.ylim(bottom=20,top=65)
plt.xticks(rotation = 90)
plt.xlabel("Model Name",fontsize=20)
plt.ylabel("Age",fontsize=20)
plt.plot(df3['Model'],df3['Age'],label='Range',c='r')
```
